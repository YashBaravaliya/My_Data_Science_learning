### 1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

A Class is a blueprint or a template for creating objects. A Class defines a set of attributes and methods that are common to all objects created from that Class. Classes can be thought of as categories, and objects as instances of those categories.

An Object is an instance of a Class, created at runtime. Objects have their own attributes and methods, which are defined by the Class. Objects can interact with each other through methods and attributes, and can perform tasks and store data.

```python
class Car:
    def __init__(self, make, model, color, year):
        self.make = make
        self.model = model
        self.color = color
        self.year = year

    def start(self):
        print("Car started")
    
    def stop(self):
        print("Car stopped")
    
    def drive(self):
        print("Car is driving")
```

```python
car1 = Car("Toyota", "Camry", "Red", 2022)
car2 = Car("Honda", "Civic", "Blue", 2021)

car1.start()
car2.start()
```

### 2. Name the four pillars of OOPs.

The four pillars of Object-Oriented Programming (OOP) are:

1. Abstraction: It refers to the process of hiding the implementation details and exposing only the necessary information to the users.

2. Encapsulation: It refers to wrapping the data (instance variables) and behavior (methods) within a single unit or object.

3. Inheritance: It refers to the mechanism of acquiring the properties and behaviors of an existing class into a new class.

4. Polymorphism: It refers to the ability of a single function or operator to perform different actions depending on the type of data it operates on.

### 3. Explain why the __init__() function is used. Give a suitable example.

The __init__ function is used in Python as a constructor for classes. It is called when an object of a class is created and is used to initialize the attributes of the object.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person1 = Person("yash", 32)
person2 = Person("om", 28)

print(person1.name) 
print(person2.age) 
```

### 4. Why self is used in OOPs?

In Object-Oriented Programming (OOPs), self is a reference to the instance of the class and is used to access its attributes and methods. The self parameter is automatically passed to each method in the class when the method is called. It is convention to use self as the first parameter in a class method, but any valid Python identifier can be used. The use of self allows each instance of a class to have its own unique set of attributes and methods, making it an essential part of OOPs.

### 5. What is inheritance? Give an example for each type of inheritance.

Inheritance is a mechanism in Object-Oriented Programming (OOP) that allows a new class to inherit properties and behaviors from an existing class. This provides a way to reuse code, simplify the code maintenance and increase the overall code readability. There are several types of inheritance, they are:

1. Single inheritance: In single inheritance, a class inherits properties from a single parent class. 

For example, consider a class 'Vehicle' and a class 'Car' that inherits properties from the class 'Vehicle'.

2. Multiple inheritance: In multiple inheritance, a class inherits properties from more than one parent class. 

For example, consider a class 'Person' and a class 'Employee' that inherits properties from both 'Person' and 'Department' class.

3. Multi-level inheritance: In multi-level inheritance, a class inherits properties from a parent class, and then the child class inherits from the derived class. 

For example, consider a class 'Animal', a class 'Mammal' that inherits from the 'Animal' class and a class 'Human' that inherits from the 'Mammal' class.

4. Hierarchical inheritance: In hierarchical inheritance, multiple child classes inherit properties from a single parent class. 

For example, consider a class 'Shape' and classes 'Triangle', 'Square' and 'Rectangle' that inherits from the class 'Shape'.

5. Hybrid inheritance: It is a combination of multiple inheritance types. 

For example, consider a class 'Animal' and a class 'Mammal' that inherits from the 'Animal' class, and classes 'Human' and 'Dog' that inherit from the 'Mammal' class.
